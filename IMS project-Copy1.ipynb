# Add new item into inventary

```python

    Add=input("do you want to add new record(Y/N)")
    if(Add=="Y"):
        id=int(input("enter product id:"))
        name=input("enter product name:")
        cat=input("enter product category:")
        prize=int(input("enter product prize:"))
        qty=int(input("enter product quantity:"))
        dis=int(input("enter product discount(%):"))
        records[id]={}
        records[id]["prod_name"]=name
        records[id]["category"]=cat
        records[id]["prize"]=prize
        records[id]["quantity"]=qty
        records[id]["discount"]=dis
        print("********************************************************************")
        print("***********         RECORD ADDED SUCCESSFULLY        ***************")
    else:
        print("********************************************************************")
        print("***********          NO  RECORD ADDED                ***************")
```
